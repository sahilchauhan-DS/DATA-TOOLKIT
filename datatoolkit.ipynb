#6.What are the differences between NumPy arrays and Python lists?
->NumPy arrays are optimized for numerical computing and large datasets, offering speed and memory efficiency, while Python lists provide flexibility and versatility for general-purpose tasks.
##Data Type Homogeneity:
NumPy arrays are designed to store elements of the same data type, which enhances efficiency in numerical operations and memory usage. In contrast, Python lists can accommodate elements of diverse data types within the same list.
##Memory Efficiency:
NumPy arrays generally consume less memory compared to Python lists, particularly when dealing with large datasets, due to the homogeneous data type storage and contiguous memory allocation. Python lists, with their flexibility to store different data types, often require more memory overhead.
##Computational Performance:
NumPy arrays excel in numerical computations, offering significantly faster operations compared to Python lists. This performance advantage stems from the optimized C implementation of NumPy and the ability to perform vectorized operations on arrays.
##Functionality:
NumPy provides a rich set of built-in functions and tools specifically designed for array manipulation, mathematical operations, and scientific computing. Python lists, while versatile, lack these specialized functionalities.

#15.How does NumPy handle multidimensional arrays?
->NumPy's core is the ndarray, designed for efficient handling of multidimensional arrays. These arrays, also known as matrices or tensors, are structured grids of elements of the same data type, organized along axes (dimensions).
Internally, NumPy stores ndarray data in a contiguous block of memory, which is treated as a 1D array. Metadata, such as shape and strides, defines how this 1D data is accessed as a multidimensional array. The shape attribute specifies the size of each dimension, while strides define the number of bytes to step in each dimension when traversing the array. This approach allows for flexible reshaping and slicing operations without moving the underlying data.

```python
#4.How do you calculate the correlation matrix using Seaborn and visualize it with a heatmapHow do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?
#ans
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [5, 4, 3, 2, 1],
    'C': [2, 3, 4, 5, 6],
    'D': [1, 3, 2, 4, 5]
}

df = pd.DataFrame(data)

# 1. Calculate the correlation matrix
corr_matrix = df.corr()

# 2. Visualize the correlation matrix with a heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar=True)

# Add title
plt.title('Correlation Matrix Heatmap')

# Show the plot
plt.show()

```

```python
#6.Create a DataFrame and add a new column based on an existing column?
#ans
import pandas as pd

# 1. Create a sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40]
}

df = pd.DataFrame(data)

# 2. Add a new column based on the 'Age' column
# For example, create a new column 'Age Group' based on the 'Age'
df['Age Group'] = df['Age'].apply(lambda x: 'Young' if x < 30 else 'Adult')

# 3. Display the updated DataFrame
print(df)
```

```python
#8.Create a line plot with multiple lines using Matplotlib?
#ans
import matplotlib.pyplot as plt

# Sample data for multiple lines
x = [0, 1, 2, 3, 4, 5]
y1 = [0, 1, 4, 9, 16, 25]  # Line 1
y2 = [0, 1, 2, 3, 4, 5]    # Line 2
y3 = [0, 2, 4, 6, 8, 10]   # Line 3

# Create the plot
plt.plot(x, y1, label='y = x^2', color='blue')   # Plot for Line 1
plt.plot(x, y2, label='y = x', color='green')    # Plot for Line 2
plt.plot(x, y3, label='y = 2x', color='red')     # Plot for Line 3

# Add labels and title
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Multiple Lines Plot')

# Show legend
plt.legend()

# Display the plot
plt.show()
```

```python
#9.Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold
#ans
import pandas as pd

# 1. Create a sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, 35, 40, 22]
}

df = pd.DataFrame(data)

# 2. Set a threshold for the 'Age' column (e.g., Age > 30)
threshold = 30

# 3. Filter rows where 'Age' is greater than the threshold
filtered_df = df[df['Age'] > threshold]

# 4. Display the filtered DataFrame
print(filtered_df)
```

```python
#11.Perform matrix multiplication using NumPy?
#ans
import numpy as np

# 1. Create two NumPy arrays (matrices)
A = np.array([[1, 2], [3, 4]])  # 2x2 matrix
B = np.array([[5, 6], [7, 8]])  # 2x2 matrix

# 2. Perform matrix multiplication using np.dot()
result = np.dot(A, B)

# Alternatively, you can use the @ operator:
# result = A @ B

# 3. Print the result
print("Matrix A:\n", A)
print("Matrix B:\n", B)
print("Result of A * B:\n", result)
```
